```python
try:
    os.chdir('/home/wifi/Downloads/simple/Idea-validation-ip-ccn/TestLogs_01032108/')
except FileNotFoundError :
    pass
filesList = os.listdir()
filesList.sort()
#print(filesList)
for fileName in filesList:
    index = 44
    _offset = 0
    if int(fileName[0]) > 6:
        index = 41
    if int(fileName[0]) > 6 or (int(fileName[0]) == 2 and int(fileName[6]) < 4 and fileName[4] == 's'):
        _offset = 1
    try:
        with open(fileName, 'r') as fp :
            _avgTpt = 0
            _avgSiz = 0
            _maxSiz = 0
            _curSiz = 0
            _curLat = 0
            _avgLat = 0
            _avgJmp = 0
            _curJmp = 0
            _maxJmp = 0
            _minJmp = 0xFF
            _stdJmp = 0
            _numRcd = 0
            _timTrk = 0
            _avgPdr = 0
            _srtNum = 0
            _numSec = 0
            for line in fp.readlines():
                try:
                    if line[index] in ('_', ' '):
                        continue
                except IndexError:
                    pass
                try:
                    if line[18] != _timTrk :
                        if _timTrk :
                            _numSec += 1
                            _avgPdr = (_avgPdr * (_numSec - 1) +  _numRcd - _srtNum) / _numSec
                            _srtNum = _numRcd
                        _timTrk = line[18]
                    if int(line[index]) == (6 + _offset) :
                        _curSiz = int(line[-4:]) - 16
                        if _curSiz > _maxSiz :
                            _maxSiz = _curSiz
                        _numRcd += 1
                        _avgSiz = (_avgSiz * (_numRcd - 1) + _curSiz) / _numRcd
                    elif int(line[index]) == (7 + _offset) :
                        _curJmp = line[45:].count(',')
                        if _curJmp > _maxJmp :
                            _maxJmp = _curJmp
                        if _curJmp < _minJmp :
                            _minJmp = _curJmp                        
                        _avgJmp = (_avgJmp * (_numRcd - 1) + _curJmp) / _numRcd
                    elif int(line[index]) == (8 + _offset) :
                        _hr = int(line[-15:-14])
                        _mn = int(line[-13:-11])
                        _se = int(line[-10:-8])
                        _mr = int(line[-7:])
                        if not _hr and not _mn :
                            _curLat = _se * 1000 + _mr / 1000
                        else :
                            _curLat = _hr * 60 * 60 * 1000000 + _mn * 60 * 1000000 + _se * 1000 + _mr / 1000
                        _avgLat = (_avgLat * (_numRcd - 1) + _curLat) / _numRcd
                except IndexError:
                    pass
        1/_avgSiz
        _avgSiz = round(_avgSiz, 3)
        _avgLat = round(_avgLat, 3)
        _avgTpt = round(_avgSiz * _avgPdr/1000, 3)
        _avgPdr = round(_avgPdr, 3)
        _avgJmp = round(_avgJmp)
        _stdJmp = (_maxJmp + _minJmp)/2 - _avgJmp
        _numOfNodes = int(fileName[0]) * int(fileName[0])
        _mode = 'sdn' if fileName[4] == 's' else 'mesh'
        _loss = None
        try:
            _loss = int(fileName[6:8])
        except ValueError :
            _loss = int(fileName[6])
        print("{0},{1:d},{2:},{3:d},{4:.3f},{5:.3f},{6:.3f},{7:.3f},{8:d},{9:d},{10:d},{11:.3f},{12:d}".format(fileName[:3], _numOfNodes, _mode, _loss, _avgTpt, _avgLat, _avgSiz, _avgPdr, _avgJmp, _minJmp, _maxJmp, _stdJmp, _maxSiz))
    except ZeroDivisionError:
        _numOfNodes = int(fileName[0]) * int(fileName[0])
        _mode = 'sdn' if fileName[4] == 's' else 'mesh'
        _loss = None
        try:
            _loss = int(fileName[6:8])
        except ValueError :
            _loss = int(fileName[6])
        print("{0},{1:d},{2:},{3:d}".format(fileName[:3], _numOfNodes, _mode, _loss))
        #print('\n')
        #raise ZeroDivisionError
        pass
    except ValueError:
        print(fileName, end=' ')
        print(line)
        raise ValueError
    #break
```
